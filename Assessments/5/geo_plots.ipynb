## Shapely: Point, LineString, Polygon

```python
from shapely.geometry import Point, LineString, Polygon

p=Point((3,4))
print(p.geom_type) # Point which has no area or length
test=LineString([Point(0,0), Point(3,4), Point(6,0)])

q=LineString([ (0,0), (3,4), (6,0) ])
print(q.geom_type) # LineString which has no area
print(q.length) # 10

r=Polygon(q)
print(r.geom_type) # Polygon
print(r.length) # 16
print(r.area) # 12
```

## Shapely: Polygons & MultiPolygon

```python
"""
Using Shapely, define an instance of a
MultiPolygon. Print it out. Confirm
the geometry type, length, and area
"""
from shapely.geometry import Polygon, MultiPolygon

p1 = Polygon( [ (0,0), (3,4), (6,0) ] )
p2 = Polygon( [ (0,0), (3,-4), (6,0) ] )
p3 = Polygon( [ (0,0), (-3,4), (-6,0) ] )

m1=MultiPolygon([ p1, p2, p3 ])

print(m1.geom_type) # MultiPolygon
print(m1.length, m1.area) # 48, 36
```
